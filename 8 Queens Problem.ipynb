```python
import chess as ch
import pandas as pd
import random
import time
```

```python
def calculate_movement_collapse(vezir_parmağı=[]):
    collapse=0 
    for i in range(len(vezir_parmağı)):
        for j in range(i+1,len(vezir_parmağı)):
            if(vezir_parmağı[j]==vezir_parmağı[i] or abs(vezir_parmağı[i]-vezir_parmağı[j])==abs(i-j)):
                collapse=collapse+1 
    return (collapse)
```

```python
def calculate_all_movements(queens=[]):
    #b=0
    collapses=[]
    for l in range(len(queens)):
        temp=queens.copy()
        for k in range(len(queens)):
            if(k!=queens[l]):
                temp[l]=k
               # import ipdb; ipdb.set_trace()
                collapses.append(calculate_movement_collapse(vezir_parmağı=temp))
            else:
                collapses.append(int(len(queens)*(len(queens)-1)/2)) #vezirlerin şu anki bulunduğu konuma 
                #en yüksek değeri atıyoruz ki seçim işleminde elensinler
        #collapses[queens[l]+len(queens)*b]="Q"
        #b+=1
    return collapses
```

```python
def generate_randomly(vezir_parmağı):
    vezir_parmağı=[]
    import random as r
    ra=8
    for x in range(ra):
        vezir_parmağı.append(r.randint(0,ra-1))
    return vezir_parmağı
```

```python
def illustrate_board(vezir_parmağı,ch): #vezir_parmağı, vezirlerin bulunduğu satır konumlarını, dizinin indeksi de sütun konumlarını ifade etmektedir
    board = ch.Board()
    board.clear()
    b=0
    print(" ",end="   ")
    
    for i in range(len(vezir_parmağı)):
        print (i,end="     ")
    print("\n")
    for i in vezir_parmağı:
        board.set_piece_at(square=8*i+b,piece=ch.Piece.from_symbol('q')) 
        #board.set_piece_at(square=i+8b,piece=chess.Piece.from_symbol('q')) olursa vezirler kendi satırında hareket eder
        b+=1
    for i in range(len(vezir_parmağı)):
        print(i,end=" ")
        for j in range(len(vezir_parmağı)):
            if(vezir_parmağı[j]==i):# if(vezir_parmağı[j]==i) de i sütünu j satırı dolaşıyor ve j deki değişim satıdaki yukarı aşağı hareket olarak çizdirilecektir
                # if(vezir_parmağı[i]==j) olursa vezilerdeki yer değişimi kendi satırında haraket değişimi ile çizdirilirdi
                print("  Q  ",end=" ")
            else:
                print("  *  ",end=" ")
        print("\n")
    
    return board
```

```python
def print_collapses(collapses,number_of_queens):
    print("    ",end="")
    for i in range(number_of_queens):
        print(' {:>2} '.format(i),end="")
    print("\n    ",end="")
    for i in range(number_of_queens):
        print(" {:>2} ".format("-"),end="")
    print("\n")
    for i in range(number_of_queens):
        print(i,"|",end=" ")
        for j in range(number_of_queens):
            print(' {:>2} '.format(collapses[i+number_of_queens*j]),end="")
        print("\n")
    #print("collapses dizisi: ",collapses,end="")
```

```python
def make_movement(queens=[]):
    
    collapses=calculate_all_movements(queens)
    current_min_collapse=min(collapses)
    current_min_collapse_index=collapses.index(current_min_collapse)
        
    num_of_Equal_collapses=0
    Equal_collapses_indexes=[]
    for i in range(len(collapses)):
        if(collapses[i]==current_min_collapse):
            Equal_collapses_indexes.append(i)
            num_of_Equal_collapses+=1
    rnd_index=random.randint(0,num_of_Equal_collapses-1)#Eşit çakışma yaratacak vezir hareketleri içerisinden rastgele seçilir

    current_min_collapse_index=Equal_collapses_indexes[rnd_index]
    
    
    coloumn=int((current_min_collapse_index-current_min_collapse_index%len(queens))/len(queens)) 
    row=current_min_collapse_index%len(queens)
        # vezirin hareket edeceği satır ve sütun

    print("\nEn küçük çakışma değerine sahip index: ",current_min_collapse_index)
    print("En küçük çakışma değeri : ",current_min_collapse)
    print("İlgili vezir ",coloumn," numaralı sütun ",row," numaralı satıra gidecek")

    queens[coloumn]=row
    #Veziri hareket ettirme
    return current_min_collapse
```
